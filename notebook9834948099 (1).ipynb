```python
import pandas as pd
df = pd.read_csv("/kaggle/input/olympic-games/summer.csv")
df = pd.read_csv("/kaggle/input/olympic-games/winter.csv")
print(df.columns)
print(df.head())
# Standardize country column
df.rename(columns={'Team':'Country'}, inplace=True)

# Gender column might be 'Sex'
df.rename(columns={'Sex':'Gender'}, inplace=True)

# Total Medals: only for medal-winning rows
df['Gold'] = (df['Medal'] == 'Gold').astype(int)
df['Silver'] = (df['Medal'] == 'Silver').astype(int)
df['Bronze'] = (df['Medal'] == 'Bronze').astype(int)
df['TotalMedals'] = df['Gold'] + df['Silver'] + df['Bronze']
```
